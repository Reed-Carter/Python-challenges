```python
# manipulating strings

name = "bob dylan"

# UPPER/LOWER CASE
# ---------------------------------------
cap_name = name.capitalize()        # the first word (john) is capitalized
title_name = name.title()           # first letter of all words are capitalized
upper_name = name.upper()           # the entire str is capitalized
lower_name = upper_name.lower()     # back down to lower case
print(f"capitalize(): {cap_name}")
print(f"title(): {cap_name}")
print(f"upper(): {upper_name}")
print(f"lower(): {lower_name}")


# REMOVING WHITESPACES
# ---------------------------------------
original = "   Meanwhile, far away in another part of town   \n"
print(f"Original: '{original}'")        # notice that this str has an endline or ENTER (\n) charater at the end
print(f"new: '{original.strip()}'")     # notice the enter is removed
print(f"left: '{original.lstrip()}'")     # removing leading whitespaces, stands for left strip
print(f"right: '{original.rstrip()}'")     # removing trailing whitespaces, stands for right strip

# FIND and REPLACE
# ---------------------------------------

lyrics = """
Here comes the story of the Hurricane
The man the authorities came to blame
For somethin' that he never done
"""
print(lyrics)

hurricane_index = lyrics.find("Hurricane")
print(f"Starting Index location of 'Hurricane in lyrics: {hurricane_index}")

# .replace(search, replace) replaces a substring with another substring
# .replace takes two arguments: first is the search string, and the second is what you like to replace with
robin_lyrics = lyrics.replace("Hurricane", "Robin")     # replace Hurricane with Robin
print(robin_lyrics)
```

```python
# classes 

# a globally scoped sequence
book_id_sequence = 1000

class Book:
    def __init__(self, title, author, genre):
        # access the global variable inside a class instance
        global book_id_sequence             # tell python to use the globally defined variable
        self.book_id = book_id_sequence     # take the current value of id seq
        book_id_sequence += 1               # increment the sequence globally
        self.title = title
        self.author = author
        self.genre = genre
    
    def have_you_read(self):
        # add printing of the book_id here
        print(f"Have you read {self.title} by {self.author}? (book id: {self.book_id})")

#create instances for the book class, the first string is the title, then the author, then the genre as defined in the class def
runaway_ralph = Book("Runaway Ralph", "Beverly Cleary", "Children's Novel")
animorphs = Book("The Invasion", "K. A. Applegate", "YA SciFi")

#this is how u call the class
runaway_ralph.have_you_read()
runaway_ralph.author
```

```python
#recursion example - function is used within its own function

#takes a number and returns the sum of all numbers up to that number
def add_recursion(num):
    if num > 0:
        result = num + add_recursion(num - 1)
        print(result)
    else:
        result = 0
    return result
print(f"lets check the value for 3: {add_recursion(3)}")
```

```python
#typehints and doc-strings example

new_flavors = ["lobster", "onion", "buttered popcorn"]
def new_selters(seltzer_flavors: str) -> str:
    """ Puts each value of an iterable into sentence.

    Args:
        seltzer_flavors (list): a list of new flavors, as strings

    Returns: 
        str: each flavor formatted into a sentence
    """
    formatted_flavors = ""
    for flavor_index in range(len(seltzer_flavors) - 1):
        formatted_flavors += f"{seltzer_flavors[flavor_index]}, "
    formatted_flavors += f"and {seltzer_flavors[-1]}."
    return f"The new flavors are {formatted_flavors}"

print(new_selters(new_flavors))
```

```python
#-------------------column transformations-------------------
#changes the values in the 'codeshare' column to 'Yes' and 'No'
# define our mapper funcion
def map_codeshares(value):
    if value == 'Y':
        return 'Yes'
    elif pd.isnull(value):
        return 'No'
    else:
        return 'N/A'

# run on transformation function on the codeshare column
# notice how we are passing our mapper function to .map()
# map() automatically calls this function over each value of the codeshare column
routes["codeshare"] = routes["codeshare"].map(map_codeshares)
print(routes.sample(n=20))

#changes the values of various abbreviations to their long name:
# lookup the airline name using a lookup dict
def decode_airline(value:str):
    AIRLINE_MAP = {
        'AA': 'American Airlines', 
        'AS': 'Alaska Airlines', 
        'DL': 'Delta Air Lines',
        'UA': 'United Airlines', 
        'WN': 'Southwest Airlines',
    }
    if value in AIRLINE_MAP:
        return AIRLINE_MAP[value]
    else:
        return 'Other'
# decode airline names and assign to a new column
routes['airline_name'] = routes.airline.map(decode_airline)
# print only decoded flights (not Others)
print(routes.loc[routes.airline_name != 'Other'].sample(n=10))

#combining the values of various columns into a new column called 'route_key' where 'airline','src', and 'dest' are the columns
def encode_route_key(row):
    # a dataframe row is passed. access columns with row.column_name
    route_key = f"{row.airline}-{row.src}-{row.dest}"
    return route_key
# apply a function over entire row values
# set axis=1 to apply function over rows. axis=0 would apply over columns
routes['route_key'] = routes.apply(encode_route_key, axis=1)
print(routes)

#make 2 new columns 'nonstop' and 'loop' depending on if there are no stops and if src is also the dest.
# a single function that returns two columns at once
def encode_route_type(row):
    # nonstop: stops = 0
    if row.stops == 0:
        nonstop = True
    else:
        nonstop = False
    # loop: src == dest
    if row.src == row.dest:
        loops = True
    else:
        loops = False
    # return as a tuple
    return (nonstop, loops)
# apply a function over row values and
# unpack multiple return column values by using zip()
routes['nonstop'], routes['loop'] = zip( *routes.apply(encode_route_type, axis=1) )
print(routes.sample(n=10))
# print flights with stops
print("\nflights with multiple stops:")
print(routes.loc[routes.nonstop == False].sample(n=5))
# print the looping flight
print("\nlooping flights:")
print(routes.loc[routes.loop == True])
```
